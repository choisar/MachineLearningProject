```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=4)
lm_labels='lad'
lm_features = (cabbage_weather.drop('lad', axis=1)).columns

for i, feature in enumerate(lm_features):
  row = i//4
  col = i%4
  sns.regplot(x=lm_labels, y=feature, data=cabbage_weather, ax=axs[row][col])
```

```python
data=cabbage_weather.drop('lad', axis=1)
# data=data.drop('일시', axis=1)
data.tail()
```

```python
label = cabbage_weather[['lad']]
label.tail()
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# 데이터를 훈련 세트와 테스트 세트로 나눔
X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=0.2, random_state=42)

# XGBoost 모델 초기화 및 학습
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror')
xg_reg.fit(X_train, y_train)

# 테스트 세트에 대한 예측 수행
y_pred = xg_reg.predict(X_test)

# 평균 제곱 오차(Mean Squared Error) 계산
mse = mean_squared_error(y_test, y_pred)
xgboost_r2 = r2_score(y_test, y_pred)
rmse=np.sqrt(mse)
print("Mean Squared Error:", mse)
print("R-squared:", xgboost_r2)
print('rmse : ', rmse)
```

#최적의 파라미터

```python
from sklearn.model_selection import GridSearchCV

def find_best_params(model, param_grid, X_train, y_train, X_test, y_test):
    # GridSearchCV 모델 생성
    grid_model = GridSearchCV(model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', refit=True)

    # GridSearchCV 모델 학습
    grid_model.fit(X_train, y_train)

    # 최적의 파라미터 출력
    best_params = grid_model.best_params_
    print("최적의 파라미터:", best_params)

    # 테스트 데이터로 예측
    predictions = grid_model.predict(X_test)

    # 성능 평가
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    print("Mean Squared Error:", mse)
    print("Root Mean Squared Error:", rmse)

    return grid_model
```
