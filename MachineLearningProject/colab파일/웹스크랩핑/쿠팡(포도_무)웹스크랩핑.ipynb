```python
import requests
import re
from bs4 import BeautifulSoup

keyword='고구마'
pageNum=1
proList=[]

def findGoguma(keyword, pageNum, proList):
    url='https://www.coupang.com/np/search?q='+keyword+'&channel=user&component=&eventCategory=SRP&trcid=&traid=&sorter=scoreDesc&minPrice=&maxPrice=&priceRange=&filterType=&listSize=60&filter=&isPriceRange=false&brand=&offerCondition=&rating=0&page='+str(pageNum)+'&rocketAll=false&searchIndexingToken=1=9&backgroundColor='
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36", "Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"}

    res = requests.get(url, headers=headers)
    res.raise_for_status() # 웹페이지의 상태가 정상인지 확인

    soup = BeautifulSoup(res.text, "lxml") # 가져온 HTML 문서를 파서를 통해 BeautifulSoup 객체
    ul_tag = soup.select_one('#productList')  # id가 productList인 ul 태그 선택

    if ul_tag:
        li_tags = ul_tag.find_all('li', class_=lambda x: x and 'ad-badge-text' not in x.split())


        for li_tag in li_tags:
            p_t = li_tag.select_one('div.name').text
            p_t = re.sub('\n', '', p_t)
            p_t = p_t.lstrip()
            proList.append(p_t)
            print('[' + str(len(proList)) + ']', p_t)
            print('='*50)

    else:
        print("해당하는 상품이 없습니다.")



while(len(proList)<=300):
    proCnt=findGoguma(keyword, pageNum, proList)
```

```python
import pandas as pd

df=pd.DataFrame(proList, columns=['고구마'])
df
```

```python
df['고구마'] = df['고구마'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", "")
df[:5]
```

```python
from konlpy.tag import Okt

for i in df:
  okt = Okt()
  print(okt.morphs(i))
  break
```

```python
okt = Okt()

def tokenize(text):
    tokens = okt.pos(text)
    return [morpheme for morpheme, pos in tokens]

df['형태소'] = df['고구마'].apply(tokenize)
df
```

```python
from collections import Counter


all_tokens = [token for tokens_list in df['형태소'] for token in tokens_list]
token_counts = Counter(all_tokens)

# 결과 출력
print("형태소 등장 횟수:")
for token, count in token_counts.items():
    print(f'{token}: {count}번')
```
