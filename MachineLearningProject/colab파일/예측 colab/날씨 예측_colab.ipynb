#지역별 전체 날씨 데이터

```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/날씨예측/공용날씨/날씨 전체 csv.csv')
df.tail()
```

```python
data = df[['지점','일시', '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '평균최저기온(°C)', '평균최고기온(°C)',
       '평균 해면기압(hPa)', '합계 강수량(mm)', '일 최다 강수량(mm)', '평균 상대습도(%)', '합계 일조시간(hr)', '일조율(%)', '합계 일사량(MJ/m2)', '평균 풍속(m/s)',
       '최다 풍향(16방위)', '합계 소형 증발량(mm)', '합계 대형 증발량(mm)']]
data.tail()
```

```python
median_values = data.median()
data = data.fillna(median_values)
data.tail()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=data.rename(columns={'지점':'branch','일시':'year', '평균기온(°C)':'avg_temp', '최저기온(°C)':'min_temp', '최고기온(°C)':'max_temp', '평균최저기온(°C)':'avg_min_temp', '평균최고기온(°C)':'avg_max_temp', '평균 해면기압(hPa)':'avg_sea_pre', '합계 강수량(mm)':'tot_rain', '일 최다 강수량(mm)':'day_max_rain', '평균 상대습도(%)':'avg_rel_hum', '합계 일조시간(hr)':'tot_sun_time', '일조율(%)':'sun_rate', '합계 일사량(MJ/m2)':'tot_sun', '평균 풍속(m/s)':'avg_wind', '최다 풍향(16방위)':'max_wind_dir', '합계 소형 증발량(mm)':'tot_small_Eva', '합계 대형 증발량(mm)':'tot_big_Eva'})
```

```python
df_data=data[['year','branch']]
# df_data=df_data.values.reshape(-1, 1)
df_data[-5:]
```

```python
df_labels=data.drop(['year','branch'], axis=1)
df_labels.head()
```

#모델 학습

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(df_data, df_labels, test_size=0.3, random_state=62)
```

#Linear_model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

#Ridge_model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
```

#Lasso_model

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=1.0)
lasso_model.fit(X_train, y_train)
```

# RandomForest_model

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
```

# DecsionTree_model

```python
from sklearn.tree import DecisionTreeRegressor

tree_model = DecisionTreeRegressor()
tree_model.fit(X_train, y_train)
```

# KNeighbors_model

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor()
knn_model.fit(X_train, y_train)
```

# MLPR_model

```python
from sklearn.neural_network import MLPRegressor
mlp_model = MLPRegressor()
mlp_model.fit(X_train, y_train)
```

# 모델 평가

```python
from sklearn.metrics import mean_squared_error, r2_score

# Ridge 모델 평가
ridge_y_pred = ridge_model.predict(X_test)
ridge_mse = mean_squared_error(y_test, ridge_y_pred)
ridge_r2 = r2_score(y_test, ridge_y_pred)
ridge_rmse = np.sqrt(ridge_mse)
print("Ridge Model:")
print("r2_score:", ridge_r2)
print("MSE:", ridge_mse)
print(f"RMSE: {ridge_rmse}")
print('-'*50)

# Lasso 모델 평가
lasso_y_pred = lasso_model.predict(X_test)
lasso_mse = mean_squared_error(y_test, lasso_y_pred)
lasso_r2 = r2_score(y_test, lasso_y_pred)
lasso_rmse = np.sqrt(lasso_mse)
print("Lasso Model:")
print("r2_score:", lasso_r2)
print("MSE:", lasso_mse)
print(f"RMSE: {lasso_rmse}")
print('-'*50)

# DecisionTreeRegressor 모델 평가
tree_y_pred = tree_model.predict(X_test)
tree_mse = mean_squared_error(y_test, tree_y_pred)
tree_r2 = r2_score(y_test, tree_y_pred)
tree_rmse = np.sqrt(tree_mse)
print("DecisionTreeRegressor Model:")
print("r2_score:", tree_r2)
print("MSE:", tree_mse)
print(f"RMSE: {tree_rmse}")
print('-'*50)

# 선형 회귀 모델
linear_y_pred = lr.predict(X_test)
linear_mse = mean_squared_error(y_test, linear_y_pred)
linear_r2 = r2_score(y_test, linear_y_pred)
linear_rmse = np.sqrt(linear_mse)
print("Linear Regression Model:")
print("r2_score:", linear_r2)
print("MSE:", linear_mse)
print(f"RMSE: {linear_rmse}")
print('-'*50)

# 랜덤 포레스트 회귀 모델
rf_y_pred = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_r2 = r2_score(y_test, rf_y_pred)
rf_rmse = np.sqrt(rf_mse)
print("Random Forest Regression Model:")
print("r2_score:", rf_r2)
print("MSE:", rf_mse)
print(f"RMSE: {rf_rmse}")
print('-'*50)

# k-최근접 이웃 회귀 모델
knn_y_pred = knn_model.predict(X_test)
knn_mse = mean_squared_error(y_test, knn_y_pred)
knn_r2 = r2_score(y_test, knn_y_pred)
knn_rmse = np.sqrt(knn_mse)
print("K-Nearest Neighbors Regression Model:")
print("r2_score:", knn_r2)
print("MSE:", knn_mse)
print(f"RMSE: {knn_rmse}")
print('-'*50)

# 다층 퍼셉트론 회귀 모델
mlp_y_pred = mlp_model.predict(X_test)
mlp_mse = mean_squared_error(y_test, mlp_y_pred)
mlp_r2 = r2_score(y_test, mlp_y_pred)
mlp_rmse = np.sqrt(mlp_mse)
print("Multilayer Perceptron Regression Model:")
print("r2_score:", mlp_r2)
print("MSE:", mlp_mse)
print(f"RMSE: {mlp_rmse}")
```

# 최적의 파라미터

```python
from sklearn.model_selection import GridSearchCV

def find_best_params(model, param_grid, X_train, y_train, X_test, y_test):
    # GridSearchCV 모델 생성
    grid_model = GridSearchCV(model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', refit=True)

    # GridSearchCV 모델 학습
    grid_model.fit(X_train, y_train)

    # 최적의 파라미터 출력
    best_params = grid_model.best_params_
    print("최적의 파라미터:", best_params)

    # 테스트 데이터로 예측
    predictions = grid_model.predict(X_test)

    # 성능 평가
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    print("Mean Squared Error:", mse)
    print("Root Mean Squared Error:", rmse)

    return grid_model
```

#교차 검증

```python
from sklearn.model_selection import cross_val_score, cross_validate

def get_avg_rmse_cv(models):
  for model in models:
    rmse_list = np.sqrt(-cross_val_score(model, df_data, df_labels, scoring='neg_mean_squared_error', cv=5))

    rmse_avg = np.mean(rmse_list)

    print(f'{model.__class__.__name__} CV RMSE 값 리스트 : {np.round(rmse_list, 3)}')
    print(f'{model.__class__.__name__} 평균 RMSE 값 : {np.round(rmse_avg, 3)}')
```

```python
models = [lr,knn_model]
get_avg_rmse_cv(models)
```
